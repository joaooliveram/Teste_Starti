```python
# Passo a passo
# Passo 1: Importar base de dados
# Passo 2: Visualizar base de dados
# Passo 3: Corrigir os erros da base de dados

import pandas as pd
tabela = pd.read_csv("dadosjusbr-remuneracoes EDITADO  - dadosjusbr-remuneracoes.csv")
#print(tabela.info())
#print("..............................")

tabela = tabela.drop(columns= ["orgao"])
#display(tabela)
#print(tabela.info())

coluna_valor = tabela["valor"]
coluna_valor = coluna_valor.str.replace(".", "")
coluna_valor = coluna_valor.str.replace(",", ".")
coluna_valor = coluna_valor.astype(float)
tabela["valor"] = coluna_valor
#display(tabela)
#print(tabela.info())

#Passo 4: fazer a soma do contrachque = base
filtro_base = tabela['categoria_contracheque'] == 'base'
valor_bruto= tabela.loc[filtro_base, "valor"].sum()
print(f"O Valor Total Bruto (categoria 'base') é: {valor_bruto}")

# RESPOSTA 1    9.266.697,78

#Passo 5: fazer a soma dos detalhes do contracheque = Retenção Teto Constitucional

filtro_retencao = tabela['detalhamento_contracheque'] == 'Retenção por Teto Constitucional'
valor_retencao = tabela.loc[filtro_retencao, "valor"].sum()
print(f"O Valor Total de Retenção (categoria 'retencao') é: {valor_retencao}")

#RESPOSTA 2  -373.820,21


tabela_salario = tabela.loc[filtro_base]

maior_valor = tabela_salario['valor'].max()
pessoas_maior_salario = tabela_salario[tabela_salario['valor'] == maior_valor]
display(pessoas_maior_salario[['nome', 'valor']])

lista_nomes = pessoas_maior_salario['nome'].tolist()
nomes_juntos = ", ".join(lista_nomes)
```
